```python
# Reading the text files
newsfile = sc.textFile("file:///data2/elisalom/sparkex/h516/assignment2/data/2012*")
```

```python
# Import required libraries
import json

# Take only the “text” part of each article
articleTexts = newsfile.map(lambda x: json.loads(x)['text'])

# Convert the text into lowercase
articleWords = articleTexts.flatMap(lambda x: x.lower().split())

# Frequency of all the words
print('Frequency of the words:', articleWords.count())
```

```python
# Remove (Filter) any word whose frequency is less than 10
articleWordsMR = articleWords.map(lambda x: (x, 1)).reduceByKey(lambda a, b: a + b)
filteredArticleWords = articleWordsMR.filter(lambda x: x[1] > 10)
```

```python
# Total size of the output data 
print('Total size file after filtration):', filteredArticleWords.count())
```

```python
# Frequency of the word congress
congress = filteredArticleWords.filter(lambda x: 'congress' in x)
print('Frequency of the word congress:', congress.collect())

# Frequency of the word london
london = filteredArticleWords.filter(lambda x: 'london' in x)
print('Frequency of the word london:', london.collect())

# Frequency of the word washington
washington = filteredArticleWords.filter(lambda x: 'washington' in x)
print('Frequency of the word washington:', washington.collect())

# Frequency of the word football
football = filteredArticleWords.filter(lambda x: 'football' in x)
print('Frequency of the word football:', football.collect())
```

```python
# maximum frequency word for each month
maxFreqWordCount = []
for i in range(12):
    monthlyData = sc.textFile('file:///data2/elisalom/sparkex/h516/assignment2/data/2012-' + "{:0>2d}".format(i + 1) + '-*')
    monthlyTexts = monthlyData.map(lambda x: json.loads(x)['text'])
    monthlyWords = monthlyTexts.flatMap(lambda x: x.lower().split())
    monthlyMR = monthlyWords.map(lambda x: (x, 1)).reduceByKey(lambda a, b: a + b)
    maxFreqWordCount.append(monthlyMR.max(lambda x: x[1]))

print('The word with maximum frequency for each month:', maxFreqWordCount)
```

```python
# Reading words from ‘2012-09-01’ text file
data2012Sep01 = sc.textFile('file:///data2/elisalom/sparkex/h516/assignment2/data/2012-09-01')
text2012Sep01 = data2012Sep01.map(lambda x: json.loads(x)['text'])
words2012Sep01 = text2012Sep01.flatMap(lambda x: x.lower().split()).map(lambda x: (x, 1)).reduceByKey(lambda a, b: a + b)

# Reading words from ‘2012-08-01’ text file
data2012Aug01 = sc.textFile('file:///data2/elisalom/sparkex/h516/assignment2/data/2012-08-01')
text2012Aug01 = data2012Aug01.map(lambda x:json.loads(x)['text'])
words2012Aug01 = text2012Aug01.flatMap(lambda x: x.lower().split()).map(lambda x: (x, 1)).reduceByKey(lambda a, b: a + b)

# List of words appeared on ‘2012-09-01’ but not on ‘2012-08-01’
wordsSepMinusAug = words2012Sep01.subtractByKey(words2012Aug01)
print('List of words appeared on ‘2012-09-01’ but not on ‘2012-08-01’:', wordsSepMinusAug.collect())
print('Count of words appeared on ‘2012-09-01’ but not on ‘2012-08-01’:', wordsSepMinusAug.count())
```

```python
# Frequency of the word ‘monsoon’ for all months
monsoonFreq = []
for i in range(12):
    monthlyData = sc.textFile('file:///data2/elisalom/sparkex/h516/assignment2/data/2012-'+"{:0>2d}".format(i + 1) + '-*')
    monthlyTexts = monthlyData.map(lambda x: json.loads(x)['text'])
    monthlyWords = monthlyTexts.flatMap(lambda x: x.lower().split())
    monthlyMR = monthlyWords.map(lambda x: (x, 1)).reduceByKey(lambda a, b: a + b)
    monsoonFreq.append(monthlyMR.filter(lambda x: 'monsoon' in x).collect())

print('Frequency of the word ‘monsoon’ for all months:', monsoonFreq)
```
