```python
# reading and stripping positive words
with open("/Users/mayam/Desktop/h516 asg4/positive.txt") as f:
    filePosit = f.readlines()
posit = [x.strip() for x in filePosit]
posit
```

```python
# reading and stripping negative words
with open("/Users/mayam/Desktop/h516 asg4/negative.txt") as f:
    fileNegat = f.readlines()
negat = [x.strip() for x in fileNegat]
negat
```

```python
# Count of positive words  
allPositiveWords = read_all_files.mapValues(lambda s: s.split()).mapValues(lambda x: list(set(x).intersection(posit))).mapValues(lambda s: len(s))
allPositiveWords.toDF().show()
```

```python
# Count of negative words 
allNegativeWords = read_all_files.mapValues(lambda s: s.split()).mapValues(lambda x: list(set(x).intersection(negat))).mapValues(lambda s: len(s))

joinRDD = allPositiveWords.join(allNegativeWords)
allNegativeWords.toDF().show()
```

```python
sortedList = joinRDD.sortByKey(ascending=True).values()
```

```python
# list of positives
pos_List = sortedList.map(lambda x: x[0]).collect()
```

```python
# list of negatives
neg_List = sortedList.map(lambda x: x[1]).collect()
```

```python
# return sentiment output
def getSentiment(lst1,lst2):
    senti = []
    for i,j in zip(lst1,lst2):
        if i > j:
            senti.append("Positive")
        elif j > i:
            senti.append("Negative")
        else:
            senti.append("Both Positive and Negative")
    return(senti)
```
