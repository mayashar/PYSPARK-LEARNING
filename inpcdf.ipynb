```python
from pyspark.sql.functions import rank, col, unix_timestamp, from_unixtime, to_timestamp
from pyspark.sql import functions as F
timeFmt = "yyyy-MM-dd"
#time_T2DAge = (F.unix_timestamp('BIRTH_DATE', format=timeFmt) - F.unix_timestamp('DOB', format=timeFmt))
df.select('BIRTH_DATE',from_unixtime(unix_timestamp('BIRTH_DATE', 'MM/dd/yyy')).alias('DOB'))
df.select('F_T2D_Diag',from_unixtime(unix_timestamp('F_T2D_Diag', 'MM/dd/yyy')).alias('FT2D'))
df.select('F_AlzDiagDay',from_unixtime(unix_timestamp('F_AlzDiagDay', 'MM/dd/yyy')).alias('F_AD_DAY'))
df.select('S_AlzDiagDay',from_unixtime(unix_timestamp('S_AlzDiagDay', 'MM/dd/yyy')).alias('S_AD_DAY'))
df.select('F_AlzDementia_DiagDay',from_unixtime(unix_timestamp('F_AlzDementia_DiagDay', 'MM/dd/yyy')).alias('FAD_DEM_DAY'))
df.select('S_AlzDementia_DiagDay',from_unixtime(unix_timestamp('S_AlzDementia_DiagDay', 'MM/dd/yyy')).alias('SAD_DEM_DAY'))
```
